**MY APPROACH TO IMPLEMENT ML MODELS**
1. In traditional way, we import classes first, then we implement initialte it and then we use model.fit() etc.
2. First, I implemented train_test_split for transformed data using TfidfVectorizer, CountVectorizer.
3. So, I have xtrain_tfidf,xtest_tfidf,ytrain_........
4. Now, I stored all the objects(Binery Classifier) in a dictionery, where the objects are belongs to a particuler family.
5. after that i intiated empty lists : accuracy_tfidf,recall_tfidf,recision_tfidf for storing metrics that obtained from models that are implemente on transformed data(TFIDF).
6. similarly, i initiated empty lists : accuracy_count,..... for storing metrics obtained from Count
7. I used for loop that iterates in ml_models dictionery which contians keys as string that denotes ml model name and value is ml model object.
8. Used Keys to acces model and implemnted .fit(), implemented metrics like accuracy_score etc and stored in a corresponding lists.
9. finally create dataframe, hich shows ml_model names as indexes and column as metric names.
10. After that created data frame and printed dataframe.

```python
naive_models = {
    "ComplementNB": ComplementNB(),
    "GaussianNB": GaussianNB()
}
neighbors_models = {
    "KNeighborsClassifier": KNeighborsClassifier(),
    "NearestCentroid": NearestCentroid(),
    # Special handling for LocalOutlierFactor later
}
svm_models = {
    "LinearSVC": LinearSVC(),
    "SVC": SVC(),
    # OneClassSVM handled separately
}
ensemble_models={"Ada Boost Classifier":ada,"Bagging Classifier":bag, "Extra Tree Classifier":extra, "Gradient Boosting Classifier":gd_boost, "Hist Gradient Boosting Classifier":hg_boosting, "Random Forest Classifier":random}
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# Split the data once for TFIDF and Count
xtrain_tfidf, xtest_tfidf, ytrain_tfidf, ytest_tfidf = train_test_split(
    featureColumnTfidf, output_column, test_size=0.2, random_state=42
)
xtrain_count, xtest_count, ytrain_count, ytest_count = train_test_split(
    featureColumnCount, output_column, test_size=0.2, random_state=42
)
```

```python
# Metrics storage
accuracy_tfidf, recall_tfidf, precision_tfidf = [], [], []
accuracy_count, recall_count, precision_count = [], [], []

# Looping through each model using tqdm
for model_name, model in tqdm(ensemble_models.items(), desc="Evaluating Models"):
    
    # --- TFIDF ---
    print(f"\n--- Printing Summary for {model_name} with TFIDF ---")
    try:
        if model_name=="Hist Gradient Boosting Classifier":
            model.fit(xtrain_tfidf.toarray(), ytrain_tfidf)
            ypred_tfidf = model.predict(xtest_tfidf.toarray())
        else:
            model.fit(xtrain_tfidf, ytrain_tfidf)
            ypred_tfidf = model.predict(xtest_tfidf)

        accuracy_tfidf.append(accuracy_score(ytest_tfidf, ypred_tfidf))
        recall_tfidf.append(recall_score(ytest_tfidf, ypred_tfidf, zero_division=0))
        precision_tfidf.append(precision_score(ytest_tfidf, ypred_tfidf, zero_division=0))

    except Exception as e:
        print(f"Error in TFIDF for model {model_name}: {e}")
        accuracy_tfidf.append(e)
        recall_tfidf.append(e)
        precision_tfidf.append(e)

    # --- Count Vectorizer ---
    print(f"\n--- Printing Summary for {model_name} with Count Vectorizer ---")
    try:
        if model_name=="Hist Gradient Boosting Classifier":
            model.fit(xtrain_count.toarray(), ytrain_count)
            ypred_count = model.predict(xtest_count.toarray())
        else:
            model.fit(xtrain_count, ytrain_count)
            ypred_count = model.predict(xtest_count)
        
        accuracy_count.append(accuracy_score(ytest_count, ypred_count))
        recall_count.append(recall_score(ytest_count, ypred_count, zero_division=0))
        precision_count.append(precision_score(ytest_count, ypred_count, zero_division=0))

    except Exception as e:
        print(f"Error in Count for model {model_name}: {e}")
        accuracy_count.append(e)
        recall_count.append(e)
        precision_count.append(e)
```

```python
print(f"RESULTS FOR TFIDF ON ENSEMBEL MODELS : {ensemble_models.keys()}")
results_for_tfidf=pd.DataFrame(data={"ACCURACY":accuracy_tfidf,"RECALL":recall_tfidf,"PRECISION":precision_tfidf},index=ensemble_models.keys())
results_for_count=pd.DataFrame(data={"ACCURACY":accuracy_count,"RECALL":recall_count,"PRECISION":precision_count},index=ensemble_models.keys())
```

**NEIGHBORS FAMILY**

```python
# --- Metrics storage ---
accuracy_tfidf, recall_tfidf, precision_tfidf = [], [], []
accuracy_count, recall_count, precision_count = [], [], []

# --- Looping through Neighbors models ---
for model_name, model in tqdm(neighbors_models.items(), desc="Evaluating Neighbors Models"):
    
    # --- TFIDF ---
    print(f"\n--- Printing Summary for {model_name} with TFIDF ---")
    try:
        model.fit(xtrain_tfidf, ytrain_tfidf)
        ypred_tfidf = model.predict(xtest_tfidf)

        accuracy_tfidf.append(accuracy_score(ytest_tfidf, ypred_tfidf))
        recall_tfidf.append(recall_score(ytest_tfidf, ypred_tfidf, zero_division=0))
        precision_tfidf.append(precision_score(ytest_tfidf, ypred_tfidf, zero_division=0))

    except Exception as e:
        print(f"Error in TFIDF for model {model_name}: {e}")
        accuracy_tfidf.append(e)
        recall_tfidf.append(e)
        precision_tfidf.append(e)

    # --- Count Vectorizer ---
    print(f"\n--- Printing Summary for {model_name} with Count Vectorizer ---")
    try:
        model.fit(xtrain_count, ytrain_count)
        ypred_count = model.predict(xtest_count)

        accuracy_count.append(accuracy_score(ytest_count, ypred_count))
        recall_count.append(recall_score(ytest_count, ypred_count, zero_division=0))
        precision_count.append(precision_score(ytest_count, ypred_count, zero_division=0))

    except Exception as e:
        print(f"Error in Count for model {model_name}: {e}")
        accuracy_count.append(e)
        recall_count.append(e)
        precision_count.append(e)

# --- Special Handling for LocalOutlierFactor (unsupervised model) ---
lof = LocalOutlierFactor(novelty=True)  # Enable novelty detection to allow separate prediction

for vectorizer_name, (xtrain, xtest, ytrain, ytest, acc_list, rec_list, prec_list) in {
    "TFIDF": (xtrain_tfidf, xtest_tfidf, ytrain_tfidf, ytest_tfidf, accuracy_tfidf, recall_tfidf, precision_tfidf),
    "Count": (xtrain_count, xtest_count, ytrain_count, ytest_count, accuracy_count, recall_count, precision_count)
}.items():
    
    print(f"\n--- Printing Summary for LocalOutlierFactor with {vectorizer_name} ---")
    try:
        lof.fit(xtrain)  # Only fit on train
        ypred = lof.predict(xtest)
        # LocalOutlierFactor returns 1 for inliers, -1 for outliers
        # Mapping 1 -> 0 (normal), -1 -> 1 (anomaly)
        ypred = (ypred == 1).astype(int)

        acc_list.append(accuracy_score(ytest, ypred))
        rec_list.append(recall_score(ytest, ypred, zero_division=0))
        prec_list.append(precision_score(ytest, ypred, zero_division=0))

    except Exception as e:
        print(f"Error in {vectorizer_name} for LocalOutlierFactor: {e}")
        acc_list.append(e)
        rec_list.append(e)
        prec_list.append(e)

# --- Create Results DataFrames ---
model_names = list(neighbors_models.keys()) + ["LocalOutlierFactor"]

print(f"\nRESULTS FOR TFIDF ON NEIGHBORS MODELS : {model_names}")
results_for_tfidf = pd.DataFrame(
    data={"ACCURACY": accuracy_tfidf, "RECALL": recall_tfidf, "PRECISION": precision_tfidf},
    index=model_names
)

print(f"\nRESULTS FOR COUNT VECTORIZER ON NEIGHBORS MODELS : {model_names}")
results_for_count = pd.DataFrame(
    data={"ACCURACY": accuracy_count, "RECALL": recall_count, "PRECISION": precision_count},
    index=model_names
)
```
